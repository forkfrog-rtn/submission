### Cleaning Data

#### Membersihkan Data `day_df`

##### Mengubah Tipe Data pada `dteday`

```python
datetime_columns = ["dteday"]

for column in datetime_columns:
    day_df[column] = pd.to_datetime(day_df[column])

    day_df.info()
```

##### Mengubah `temp` & `atemp` Menjadi Skala Celcius

```python
day_df['temp'] = day_df['temp'] * 100
day_df['atemp'] = day_df['atemp'] * 100

day_df
```

```python
temp = (day_df['temp'] - 32) * 5/9
atemp = (day_df['atemp'] - 32) * 5/9

day_df['temp'] = temp
day_df['atemp'] = atemp

day_df
```

Mengubah Tipe Data `temp` & `atemp`

```python
temp = day_df['temp'].astype(int)
atemp = day_df['atemp'].astype(int)

day_df['temp'] = temp
day_df['atemp'] = atemp

day_df
```

```python
day_df.info()
```

Mengalikan `hum` & `winspeed` dengan 100

```python
hum = day_df['hum'] * 100
windspeed = day_df['windspeed'] * 100

day_df['hum'] = hum
day_df['windspeed'] = windspeed

day_df
```

#### Membersihkan Data `hour_df`

##### Mengubah Tipe Data pada `dteday`

```python
datetime_columns = ["dteday"]

for column in datetime_columns:
    hour_df[column] = pd.to_datetime(hour_df[column])

    hour_df.info()
```

## Exploratory Data Analysis (EDA)

### Explore `day_df` & `hour_df`

```python
merge_df_inner = pd.merge(left=day_df, right=hour_df, on='instant', how='inner')
merge_df_inner
```

Di musim apa sewa sepeda paling laris?

```python
merge_df_inner.groupby('season_x').agg({
    'casual_x': 'sum',
    'registered_x': 'sum',
    'cnt_x': 'sum'
})
```

Di bulan keberapa sepeda di sewa paling banyak?

```python
merge_df_inner.groupby('mnth_x').agg({
    'casual_x': 'sum',
    'registered_x': 'sum',
    'cnt_x': 'sum'
})
```

Di jam ke berapa sepeda di sewa paling banyak?

```python
merge_df_inner.groupby('hr').agg({
    'casual_y': 'sum',
    'registered_y': 'sum',
    'cnt_y': 'sum'
})
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Di musim apa sewa sepeda paling laris?

Berdasarkan Pelanggan Biasa (Casual)

```python
merge_casual = merge_df_inner.groupby('season_x').agg({
    'casual_x': 'sum'
})
```

```python
plt.figure(figsize=(10, 5))

colors_1 = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]
sns.barplot(
    x='season_x',
    y='casual_x',
    data=merge_casual,
    palette=colors_1
)

plt.title('Jumlah Penyewaan Sepeda Berdasarkan Musim (Casual)', loc='center', fontsize=15)
plt.ylabel('Jumlah Penyewaan (Casual)')
plt.xlabel('Musim')
plt.xticks(rotation=0)
plt.show()
```

Berdasarkan Pelanggan yang Berlangganan (Regsitered)

```python
merge_registered = merge_df_inner.groupby('season_x').agg({
    'registered_x': 'sum'
})
```

```python
plt.figure(figsize=(10, 5))

colors_1 = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]
sns.barplot(
    x='season_x',
    y='registered_x',
    data=merge_registered,
    palette=colors_1
)

plt.title('Jumlah Penyewaan Sepeda Berdasarkan Musim (Registered)', loc='center', fontsize=15)
plt.ylabel('Jumlah Penyewaan (Registered)')
plt.xlabel('Musim')
plt.xticks(rotation=0)
plt.show()
```

Berdasarkan Total Gabungan Casual & Regsitered

```python
merge_cnt = merge_df_inner.groupby('season_x').agg({
    'cnt_x': 'sum'
})
```

```python
plt.figure(figsize=(10, 5))

colors_1 = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]
sns.barplot(
    x='season_x',
    y='cnt_x',
    data=merge_cnt,
    palette=colors_1
)

plt.title('Jumlah Penyewaan Sepeda Berdasarkan Musim (Casual & Registered)', loc='center', fontsize=15)
plt.ylabel('Jumlah Penyewaan')
plt.xlabel('Musim')
plt.xticks(rotation=0)
plt.show()
```
